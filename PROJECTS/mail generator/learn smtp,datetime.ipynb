### Date time module 

datetime.weekday()

Returns the day of the week as an integer:

Monday  0

Tuesday  1

Wednesday  2

Thursday  3

Friday  4

Saturday  5

Sunday  6

```python
import datetime as dt
now=dt.datetime.now()
year=now.year
day=now.day
month=now.month
hour=now.hour
week=now.weekday()
print(year,day,hour,week,month)

date_of_birth=dt.datetime(year=2006,month=8,day=21,hour=20,)
print(date_of_birth)
```
